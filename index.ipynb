# PersonamRemedies

## Explications de l'API:

Cette API nous permet d'améliorer le régime alimentaire de l'utilisateur en fonction de certains problèmes de santé. En effet celui-ci doit d'abord renseigner une pathologie afin d'avoir un numéro d'inscription (unique et fixe), qu'il reçoit par mail.
L'utilisateur se verra alors attribué un numéro de maladie :le " healthconditionsID".


```python
import requests
```

## Obtention de la liste des maladies référencées par l'API

```python
URL = "https://5jocnrfkfb.execute-api.us-east-1.amazonaws.com/PersonalRemedies/nutridigm/api/v2/"
subscriptionid = "aCM1V4eAKILAl950BA9na"
ID = {"subscriptionID" : subscriptionid}
healthconditions = requests.get(URL + "healthconditions", params= ID)
healthconditions = healthconditions.json()
# healthconditions[1].items()
healthconditions
```

#### Dans notre cas le healthConditionID = 46, l'allergie au lait

## Obtention de la liste des aliments référencés dans l'API
Ainsi qu'une information concernant leur effet sur le corps. Savoir s'il est dans le groupe des bons ou mauvais aliments pour le corps

```python
URL = "https://5jocnrfkfb.execute-api.us-east-1.amazonaws.com/PersonalRemedies/nutridigm/api/v2/"
subscriptionid = "aCM1V4eAKILAl950BA9na"
ID = {"subscriptionID" : subscriptionid}
fooditems = requests.get(URL + "fooditems", params= ID)
fooditems=fooditems.json()
fooditems
```
